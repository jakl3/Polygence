```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

import plotly.figure_factory as ff

sns.set()

df = pd.read_csv('../Data/OASIS/oasis_3.csv')
df
```

### Data Cleaning

```python
df = df.dropna(axis=1, how='all') # Drop any empty columns
df = df.dropna(axis=0, how='any') # Drop any rows with empty values
df = df.rename(columns={'id':'Freesurfer ID', 'dx1':'Diagnosis', # Rename columns
                        'TOTAL_HIPPOCAMPUS_VOLUME':'TotalHippocampusVol'})
df = df.sort_values(by=['Subject', 'Age']) # Sort the columns by subject, then age
df = df.drop_duplicates(subset='Subject', keep='first') # Keep only the first visit; this is possible because
                                                        # df is sorted by age
df = df.reset_index(drop=True) # Reset the index
```

```python
def label_disease (row):
    if row['cdr'] < 0.5:
        return 0
    elif row['Diagnosis'] == 'Cognitively normal':
        return 0
    else:
        return 1
# Labels the diagnosis numerically
df.insert(6, 'Demented', df.apply (lambda row : label_disease(row), axis=1))
df
```

```python
df.loc[(df['Diagnosis'] != 'Cognitively normal') & (df['cdr'] < .5)]
# Not sure why these are not labeled as Cognitively normal, but I will treat it as undemented.
```

## Visualization of data

```python
# Removes the categorical values from df to allow for plotting
noncat_df = df.select_dtypes(exclude=['object'])
noncat_df.head()
```

```python
# Draw bar chart
demented = df[df['Demented']==1]['M/F'].value_counts() / len(df[df['Demented']==1])
cn = df[df['Demented']==0]['M/F'].value_counts() / len(df[df['Demented']==0])

df_bar = pd.DataFrame([demented*100,cn*100])
df_bar.index = ['Demented', 'Cognitively normal']
df_bar.plot(kind='bar',stacked=True, figsize=(8,5))
plt.ylabel('%')
plt.title('Relative freq of gender vs group')
plt.legend()
# plt.savefig('images/Gender vs Diagnosis')
```

```python
for column in [e for e in noncat_df.columns if e not in ['Demented', 'cdr']]:
    fg = sns.FacetGrid(noncat_df, hue='Demented', aspect=4)
    fg.map(sns.kdeplot, column, shade=True)
    fg.add_legend()
    plt.title(f'Distribution of {column}, separated by group')
    # plt.savefig(f'images/{column} Distribution.png')
```

### Features as a function of age

```python
fig, axes = plt.subplots(8, 2, figsize=(16,50))
axes = axes.flatten()
for i, column in enumerate([e for e in noncat_df.columns if e not in ['Age', 'Demented']]):
    axes[i].set_title(f'{column} vs Age')
    axes[i].set_xlabel('Age')
    axes[i].set_xlim(40,100)
    sns.scatterplot(data=noncat_df, ax=axes[i], x='Age', y=column, hue='Demented')
# plt.savefig("images/feature_vs_age.png")
```
